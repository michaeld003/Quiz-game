```python
#basic quiz game in python
def new_game():
    guesses=[]#empty list
    correct_guesses=0
    question_number=1#to represent the first question

    for key in questions:
        print("----------------------")
        print(key)
        for i in options[question_number-1]:#reason being -1 is since 1-1=0 it will index the first line of answers only 
            print(i)
        guess=(input("Enter (A,B,C OR D)"))
        guess=guess.upper()  #if the user puts an answers in lower case will convert it in upper case
        guesses.append(guess)#adds the guesses in the empty list in guesses

        correct_guesses+=check_answer(questions.get(key),guess)#will get the answer and will also pass in the guess and will count the point if guess it right
        question_number+=1 #will display all the different allocated answers for each allocated question
    
    display_score(correct_guesses,guesses) #passing in the arguments of the right guesses and the list of guesses 
def check_answer(answer,guess):#filling in the parameters with the answer and the guess the user put in
    
    if answer==guess:
        print("Correct!")
        return 1
    else:
        print("Wrong!")
        return 0

def display_score(correct_guesses,guesses):
    print("----------------------")
    print("Results")
    print("----------------------")
    print("Answers: ",end=" ")
    for i in questions:
        print(questions.get(i),end=" ")#display all of the values in the dictionary
    print("Guesses: ",end=" ")
    for i in guesses:
        print(i,end=" ")
    print()

    score=int((correct_guesses/len(questions)*100)) #to calculate the final score
    print("Your score is : "+str(score)+"%")

def play_again():
    response=input("Do you want to play again(YES/NO) :")
    response=response.upper()
    if response=="YES":
        return True
    else:
        False
        


#creating a dictionary that has key value pairs,each key has a question and a value associated with it
questions={"Who has won the most Ballon d'or?": "A",
           "What team had won the most UCLs?": "B",
           "Which country has won the most World Cups?: ": "C",
             "Which country has won the most titles in the Euros: ":"A" 
             }
 
options=[["A.Lionel Messi","B.Cristiano Ronaldo","C. Karim Benzema","D. Johan Cruyff"],
         ["A. Fc Barcelona","B. Real Madrid","C.Bayern Munich","D.Liverpool"],
         ["A. Argentina","B. France", "C. Brazil","D. England"],
         ["A. Germany","B. Greece", "C. France","D. Portugal"]]

new_game()

while play_again():
    new_game()

print("Bye")
```
